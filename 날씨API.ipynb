## 0. 데이터 불러오기

```python
import requests
import io
import requests
import pandas as pd


url = "https://apihub.kma.go.kr/api/typ01/url/kma_sfcdd3.php?tm1=20140101&tm2=20240615&stn=0&help=0&authKey=Y6kSGWbOQwupEhlmziMLPg"
response = requests.get(url)  # GET 요청

# 데이터 공백으로 파싱하여 저장
data = pd.read_csv(io.StringIO(response.text), sep='\s+', skiprows=5, header=None)

# 주어진 콜럼 이름
data.columns = [
    "TM", "STN", "WS_AVG", "WR_DAY", "WD_MAX", "WS_MAX", "WS_MAX_TM",
    "WD_INS", "WS_INS", "WS_INS_TM", "TA_AVG", "TA_MAX", "TA_MAX_TM",
    "TA_MIN", "TA_MIN_TM", "TD_AVG", "TS_AVG", "TG_MIN", "HM_AVG",
    "HM_MIN", "HM_MIN_TM", "PV_AVG", "EV_S", "EV_L", "FG_DUR", "PA_AVG",
    "PS_AVG", "PS_MAX", "PS_MAX_TM", "PS_MIN", "PS_MIN_TM", "CA_TOT",
    "SS_DAY", "SS_DUR", "SS_CMB", "SI_DAY", "SI_60M_MAX", "SI_60M_MAX_TM",
    "RN_DAY", "RN_D99", "RN_DUR", "RN_60M_MAX", "RN_60M_MAX_TM",
    "RN_10M_MAX", "RN_10M_MAX_TM", "RN_POW_MAX", "RN_POW_MAX_TM",
    "SD_NEW", "SD_NEW_TM", "SD_MAX", "SD_MAX_TM", "TE_05", "TE_10",
    "TE_15", "TE_30", "TE_50"
]

# 마지막 한 줄 삭제
data = data.iloc[:-1]

# DataFrame을 CSV 파일로 저장
data.to_csv('./csv/weather_new.csv', index=False)

data
```
