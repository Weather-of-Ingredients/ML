```python
import pandas as pd

production=pd.read_csv('./csv/production_match_area.csv')
weather=pd.read_csv('./csv/weather_area_new.csv')
```

```python
#TM을 연도, 월만 나타나게 수정
weather['YEAR'] = weather['TM'].astype(str).str[:4].astype(int)
weather['MONTH'] = weather['TM'].astype(str).str[4:6].astype(int)
```

```python
weather
```

```python
#TA : 기온 / RN_DAY : 일 강수량 / WS : 풍속 / PA : 현지기압 / SS : 일조 / HM : 상대습도
weather_selected = weather[['YEAR', 'AREA', 'MONTH', 'RN_DAY', 'WS_AVG','PA_AVG', 'SS_DAY', 'HM_AVG', 'TA_AVG']]
weather_monthly = weather_selected.groupby(['YEAR','MONTH', 'AREA']).mean().reset_index()
weather_monthly
```

```python
weather_monthly['AREA'].unique()
```

```python
weather_monthly['AREA'] =weather_monthly['AREA'].replace('제주특별자치도', '제주도')
```

```python
production['지역'].unique()
```

```python
weather_monthly['AREA'].unique()
```

```python
merged_data = pd.merge(production, weather_monthly, how='left', left_on=['년도', '지역'], right_on=['YEAR', 'AREA'])
```

```python
merged_data[merged_data['지역']=="제주도"]
```

```python
final_data = merged_data[['YEAR', 'MONTH', '지역', '작물명', '생산량', 'RN_DAY',
       'WS_AVG', 'PA_AVG', 'SS_DAY', 'HM_AVG', 'TA_AVG']]

final_data.to_csv('final_merged_production_weather_data_new.csv', index=False)
```
